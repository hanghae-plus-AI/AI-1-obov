```python
iterable = ["a", "b", "c", "d", "e"]
pbar = tqdm(
    iterable,
    total=len(iterable),  ## ì „ì²´ ì§„í–‰ìˆ˜
    desc="Description",  ## ì§„í–‰ë¥  ì•žìª½ ì¶œë ¥ ë¬¸ìž¥
    ncols=80,  ## ì§„í–‰ë¥  ì¶œë ¥ í­ ì¡°ì ˆ
    ascii=" ----",  ## ë°” ëª¨ì–‘, ì²« ë²ˆì§¸ ë¬¸ìžëŠ” ê³µë°±ì´ì–´ì•¼ ìž‘ë™
    leave=True,  ## True ë°˜ë³µë¬¸ ì™„ë£Œì‹œ ì§„í–‰ë¥  ì¶œë ¥ ë‚¨ê¹€. False ë‚¨ê¸°ì§€ ì•ŠìŒ.
)
for c in pbar:
    pbar.set_description(
        f'Current Character "{c}"'
    )  ## ë˜ëŠ” pbar.desc = f'Current Character "{c}"'
    time.sleep(0.2)
pbar.close()
```
