```python
from tqdm import tqdm
import time

for i in tqdm(range(10)):  ## tqdm 첫 번째 인자에 순회가능한 객체 전달
    time.sleep(0.1)
```

```python
pbar = tqdm(range(10))  ## tqdm 객체 생성
for i in pbar:
    time.sleep(0.01)
pbar.close()
```

```python
iterable = ["a", "b", "c", "d", "e"]
for i in tqdm(
    iterable,
    total=len(iterable),  ## 전체 진행수
    desc="Description",  ## 진행률 앞쪽 출력 문장
    ncols=3000,  ## 진행률 출력 폭 조절
    ascii=" 😆",  ## 바 모양, 첫 번째 문자는 공백이어야 작동
    leave=True,  ## True 반복문 완료시 진행률 출력 남김. False 남기지 않음.
):
    time.sleep(0.4)
```

```python
iterable = ["a", "b", "c", "d", "e"]
pbar = tqdm(
    iterable,
    total=len(iterable),  ## 전체 진행수
    desc="Description",  ## 진행률 앞쪽 출력 문장
    ncols=80,  ## 진행률 출력 폭 조절
    ascii=" ----",  ## 바 모양, 첫 번째 문자는 공백이어야 작동
    leave=True,  ## True 반복문 완료시 진행률 출력 남김. False 남기지 않음.
)
for c in pbar:
    pbar.set_description(
        f'Current Character "{c}"'
    )  ## 또는 pbar.desc = f'Current Character "{c}"'
    time.sleep(0.2)
pbar.close()
```
