```python
import json

# JSON 파일 경로
file_path = "data.json"

# 파일 읽기
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# 읽은 데이터 출력
problem_set_ids = [problem_set["id"] for problem_set in data]
problems_grouped = [item["problems"] for item in data]
problems = problems_grouped[0]
problem_sets = [problem_set for problem_set in data]
[problem["question"] for problem in problems]
[item["type"] for item in data]
[item["paragraph"] for item in data]
```

```python
def remove_answer(problem):
    score_removed = {**problem}
    del score_removed["answer"]
    return problem, score_removed


def remove_answer_problem_set(problem_set):
    problem_set_answer_removed = {**problem_set}
    problem_set_answer_removed["problems"] = [
        remove_answer(problem)[1] for problem in problem_set["problems"]
    ]
    return problem_set, problem_set_answer_removed


remove_answer_problem_set(problem_sets[0])[1]
```

```python
def solve_problem(problem):
    problem


def get_prompt(problem, paragraph):
    return {
        "role": "user",
        "content": f"""
            ### passage
            {paragraph}{problem.get('question_plus',"")}
            ### question
            {problem['question']}
            ### choices
            1. {problem['choices'][0]}
            2. {problem['choices'][1]}
            3. {problem['choices'][2]}
            4. {problem['choices'][3]}
            5. {problem['choices'][4]}
        """,
    }
```

```python
base_prompts = [
    {
        "role": "system",
        "content": """
                You are an expert at reading and solving Korean language questions. You will be provided with a passage, and after careful consideration, you must select the correct answer. The answers will always be in the range of 1 to 5, and there is exactly one correct answer; there are no cases where there is no answer or more than one answer.

                Whether you select the correct or incorrect answer, you must provide reasoning or justification for your choice. Your reasoning must always be based on the given passage.
            """,
    },
    {
        "role": "user",
        "content": """
                ### passage
                인공지능(AI)은 컴퓨터나 컴퓨터로 제어되는 기계가 인간의 능력, 특히 컴퓨터 시스템이 학습하거나 추론하거나 문제를 해결하거나 지각하는 능력을 갖출 수 있음을 나타내는 컴퓨터 과학의 분야입니다.
                ### question
                인공지능(AI)은 무엇을 설명하는 과학 분야인가?
                ### choices
                1. 인간의 감정을 연구하는 분야
                2. 컴퓨터가 인간의 능력을 모방하는 분야
                3. 생물학적 시스템을 분석하는 분야
                4. 우주 탐사를 연구하는 분야
                5. 식물의 성장을 연구하는 분야
            """,
    },
    {
        "role": "assistant",
        "content": """
                ### 질문
                인공지능(AI)은 무엇을 설명하는 과학 분야인가?
                ### 정답번호
                2
                ### 정답지문
                컴퓨터가 인간의 능력을 모방하는 분야
                ### 해설
                지문에 따르면, 인공지능(AI)은 컴퓨터 시스템이 학습, 추론, 문제 해결, 지각 등의 인간 능력을 갖추는 것을 나타내는 과학 분야이므로 정답은 2번, **'컴퓨터가 인간의 능력을 모방하는 분야'**입니다.
                #### 오답일 때
                - 1번-인간의 감정을 연구하는 분야: 인공지능은 인간의 감정을 연구하는 것이 아니라, 컴퓨터나 기계가 인간의 능력을 모방하는 것에 초점을 맞추고 있습니다.
                - 3번-생물학적 시스템을 분석하는 분야: 생물학적 시스템을 분석하는 것은 AI의 정의에 포함되지 않으며, AI는 주로 컴퓨터 시스템과 관련된 과학 분야입니다.
                - 4번-우주 탐사를 연구하는 분야: 우주 탐사는 AI의 주요 연구 분야가 아니며, AI는 컴퓨터 시스템의 학습, 추론 등을 다룹니다.
                - 5번-식물의 성장을 연구하는 분야: 식물의 성장을 연구하는 것은 AI와 무관한 분야입니다.
            """,
    },
]


def get_api_prompt(problem, paragraph):
    result = [*base_prompts]
    result.append(get_prompt(problem, paragraph))
    return result
```

```python
problem_set = remove_answer_problem_set(problem_sets[0])[1]
problem = problem_set["problems"][1]


def solve_problem(problem, paragraph):
    prompts = get_api_prompt(problem, paragraph)
    response = client.chat.completions.create(
        model="gpt-4",
        messages=prompts,
        max_tokens=3000,
    )
    return response.choices[0].message.content
```

```python
solve_problem_response = solve_problem(problem, problem_set["paragraph"])
```

```python
def select_only_one_choice(solve_problem_response):
    choice_response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            *choice_prompt,
            {
                "role": "user",
                "content": solve_problem_response,
            },
        ],
        max_tokens=3000,
    )
    print(choice_response.choices[0].message.content)
    return int(choice_response.choices[0].message.content)
```

```python
problems, paragraph = problem_sets[0]["problems"], problem_sets[0]["paragraph"]

choices = []

for problem in problems:
    solve_res = solve_problem(problem, paragraph)
    choice_res = select_only_one_choice(solve_res)
    choices.append(choice_res)

choices
```

```python
every_choices = []
for problem_set in problem_sets:
    problems, paragraph = problem_set["problems"], problem_set["paragraph"]

    for problem in problems:
        solve_res = solve_problem(problem, paragraph)
        choice_res = select_only_one_choice(solve_res)
        every_choices.append(choice_res)

    every_choices
```

```python
answers, scores = [
    item
    for sublist in [
        [problem["answer"] for problem in problem_set["problems"]]
        for problem_set in data
    ]
    for item in sublist
], [
    item
    for sublist in [
        [problem["score"] for problem in problem_set["problems"]]
        for problem_set in data
    ]
    for item in sublist
]
answers, scores

every_choices = [int(choice) for choice in every_choices]
print(every_choices)
print(answers)


def grade_exam(choices, answers, scores):
    total_score = 0  # 점수를 0으로 시작

    # choices, answers, scores 배열을 순회하며 점수 계산
    for choice, answer, score in zip(choices, answers, scores):
        if choice == answer:  # 선택이 정답과 일치하면
            total_score += score  # 해당 점수를 더함

    return total_score


grade_exam(every_choices, answers, scores)
```
