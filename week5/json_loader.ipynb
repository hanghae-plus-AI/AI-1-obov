```python
import json

# JSON 파일 경로
file_path = "data.json"

# 파일 읽기
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# 읽은 데이터 출력
problem_set_ids = [problem_set["id"] for problem_set in data]
problems_grouped = [item["problems"] for item in data]
problems = problems_grouped[0]
[problem["question"] for problem in problems]
[item["type"] for item in data]
[item["paragraph"] for item in data]
[problem_set for problem_set in data][0]
answers, scores = [
    item
    for sublist in [
        [problem["answer"] for problem in problem_set["problems"]]
        for problem_set in data
    ]
    for item in sublist
], [
    item
    for sublist in [
        [problem["score"] for problem in problem_set["problems"]]
        for problem_set in data
    ]
    for item in sublist
]
answers, scores
```
